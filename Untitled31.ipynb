```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import necessary libraries
import pandas as pd
import numpy as np
import os
from tqdm import tqdm

# Path to your folder containing all CSV files
folder_path = '/content/drive/MyDrive/Amazon_products_sales_data/'  # Update with your folder path

# List all CSV files in the directory
all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
print(f"Found {len(all_files)} CSV files to process")
```

```python
def preprocess_dataframe(df, filename):
    """Preprocess a single dataframe and add a source column."""
    # Add source column to track which file the data came from
    df['source_file'] = filename

    # Clean price columns
    price_columns = ['discount_price', 'actual_price']
    for col in price_columns:
        if col in df.columns:
            # Remove currency symbols and commas, convert to float
            df[col] = (df[col].astype(str)
                        .str.replace('₹', '')
                        .str.replace('$', '')
                        .str.replace(',', '')
                        .str.strip())
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Clean ratings
    if 'ratings' in df.columns:
        df['ratings'] = pd.to_numeric(df['ratings'].str.extract('([0-9.]+)')[0], errors='coerce')

    # Clean number of ratings
    if 'no of ratings' in df.columns:
        df['no of ratings'] = (df['no of ratings'].astype(str)
                               .str.replace(',', '')
                               .str.extract('([0-9]+)')[0])
        df['no of ratings'] = pd.to_numeric(df['no of ratings'], errors='coerce')

    return df

# Initialize an empty list to store dataframes
all_dfs = []

# Process each CSV file
for filename in tqdm(all_files, desc="Processing files"):
    try:
        file_path = os.path.join(folder_path, filename)
        # Read the CSV file
        df = pd.read_csv(file_path, on_bad_lines='skip', low_memory=False)
        # Preprocess the dataframe
        df_processed = preprocess_dataframe(df, filename)
        all_dfs.append(df_processed)
        print(f"Processed {filename}: {len(df_processed)} rows")
    except Exception as e:
        print(f"Error processing {filename}: {str(e)}")

# Combine all dataframes
if all_dfs:
    combined_df = pd.concat(all_dfs, ignore_index=True)
    print(f"\nCombined dataset shape: {combined_df.shape}")
else:
    print("No dataframes were processed successfully.")
```

```python
# Handle missing values
print("\nMissing values before cleaning:")
print(combined_df.isnull().sum())

# Fill missing values
if 'main_category' not in combined_df.columns:
    # Extract main category from source filename
    combined_df['main_category'] = combined_df['source_file'].str.replace('.csv', '').str.replace('All ', '')

# Fill other missing values
combined_df['sub_category'] = combined_df.get('sub_category', 'Not Specified')
combined_df['ratings'] = combined_df.get('ratings', combined_df['ratings'].median())
combined_df['no of ratings'] = combined_df.get('no of ratings', 0)

# Calculate price statistics
if 'discount_price' in combined_df.columns and 'actual_price' in combined_df.columns:
    combined_df['discount_percentage'] = ((combined_df['actual_price'] - combined_df['discount_price']) /
                                         combined_df['actual_price'] * 100).round(2)
    combined_df['discount_percentage'] = combined_df['discount_percentage'].clip(0, 100)  # Cap at 100%

print("\nMissing values after cleaning:")
print(combined_df.isnull().sum())
```

```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# 2. Load the dataset
print("Loading dataset...")
file_path = '/content/drive/MyDrive/amazon_analysis/amazon_products_cleaned_final.csv'
df = pd.read_csv(file_path)

# 3. Take a smaller sample for faster processing and reset index
sample_size = 20000
print(f"Taking a sample of {sample_size} products...")
df_sample = df.sample(n=min(sample_size, len(df)), random_state=42).reset_index(drop=True)

# 4. Create combined features
print("Creating combined features...")
df_sample['combined_features'] = (
    df_sample['name'].fillna('') + ' ' +
    df_sample['main_category'].fillna('') + ' ' +
    df_sample['sub_category'].fillna('')
)

# 5. Initialize and fit TF-IDF
print("\nInitializing TF-IDF vectorizer...")
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=2000,  # Reduced number of features
    ngram_range=(1, 1),  # Just unigrams for now
    min_df=2,           # Minimum document frequency
    max_df=0.9          # Maximum document frequency
)

print("Fitting TF-IDF vectorizer...")
tfidf_matrix = tfidf.fit_transform(df_sample['combined_features'])

print(f"TF-IDF matrix shape: {tfidf_matrix.shape}")

# 6. Compute cosine similarities
print("\nComputing cosine similarities...")
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

print("Similarity matrix shape:", cosine_sim.shape)

# 7. Create indices for products using the new 0-based index
indices = pd.Series(df_sample.index, index=df_sample['name']).drop_duplicates()

# 8. Recommendation function
def get_recommendations(name, cosine_sim=cosine_sim, n_recommendations=5):
    """
    Get product recommendations based on cosine similarity
    """
    try:
        # Get the index of the product
        idx = indices[name]

        # Get pairwise similarity scores
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Sort products by similarity score
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Get the scores of the most similar products
        sim_scores = sim_scores[1:n_recommendations+1]  # Skip the first one (itself)

        # Get the product indices
        product_indices = [i[0] for i in sim_scores]

        # Return the top similar products
        return df_sample[['name', 'main_category', 'sub_category', 'ratings', 'actual_price']].iloc[product_indices]

    except KeyError:
        available_products = "\n".join([f"- {name}" for name in df_sample['name'].sample(5, random_state=42).tolist()])
        print(f"Product '{name}' not found in the sample. Here are some available products:")
        print(available_products)
        return None

# 9. Test the recommendation system
print("\nTesting the recommendation system...")
sample_product = df_sample.sample(1).iloc[0]['name']
print(f"Getting recommendations for: {sample_product}")

recommendations = get_recommendations(sample_product)
if recommendations is not None:
    print("\nTop 5 Recommended Products:")
    display(recommendations)
else:
    print("No recommendations could be generated. Please try a different product name.")
```

```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# 2. Load the dataset
print("Loading dataset...")
file_path = '/content/drive/MyDrive/amazon_analysis/amazon_products_cleaned_final.csv'
df = pd.read_csv(file_path)

# 3. Take a sample if the dataset is large
sample_size = 20000  # Adjust based on your system's memory
if len(df) > sample_size:
    print(f"Taking a sample of {sample_size} products...")
    df_sample = df.sample(n=sample_size, random_state=42).reset_index(drop=True)
else:
    df_sample = df.copy()

print(f"Working with {len(df_sample)} products.")

# 4. Check available columns
print("\nAvailable columns:")
print(df_sample.columns.tolist())

# 5. Use only existing columns for features
print("\nUsing existing columns for recommendations...")
existing_columns = ['name', 'main_category', 'sub_category']

# Verify all required columns exist
missing_columns = [col for col in existing_columns if col not in df_sample.columns]
if missing_columns:
    print(f"Warning: The following columns are missing: {missing_columns}")
    print("Using available columns only...")
    existing_columns = [col for col in existing_columns if col in df_sample.columns]
    if not existing_columns:
        raise ValueError("No valid columns found for recommendations.")

# 6. Create features from available columns
print(f"Creating features from: {existing_columns}")
df_sample['features'] = df_sample[existing_columns].fillna('').agg(' '.join, axis=1)

# 7. Initialize and fit TF-IDF
print("\nInitializing TF-IDF...")
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=2000  # Limit features to save memory
)

print("Creating TF-IDF matrix...")
tfidf_matrix = tfidf.fit_transform(df_sample['features'])

# 8. Compute cosine similarity
print("Computing cosine similarities...")
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# 9. Create indices for products
indices = pd.Series(df_sample.index, index=df_sample['name']).drop_duplicates()

# 10. Recommendation function
def get_recommendations(product_name, n=5):
    """
    Get product recommendations based on similarity

    Parameters:
    - product_name: Name of the product to get recommendations for
    - n: Number of recommendations to return

    Returns:
    - DataFrame with recommended products
    """
    try:
        # Get the index of the product
        idx = indices[product_name]

        # Get pairwise similarity scores
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Sort products by similarity score
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Get the scores of the most similar products
        sim_scores = sim_scores[1:n+1]  # Skip the first one (itself)

        # Get the product indices
        product_indices = [i[0] for i in sim_scores]

        # Return the top similar products
        return df_sample.iloc[product_indices]

    except KeyError:
        print(f"\nProduct '{product_name}' not found in the sample.")
        print("\nHere are some available products you can try:")
        display(df_sample[['name', 'main_category']].sample(3))
        return None

# 11. Test the recommendation system
print("\nTesting the recommendation system...")
sample_product = df_sample.sample(1).iloc[0]
print(f"\nGetting recommendations for: {sample_product['name']}")
print(f"Category: {sample_product['main_category']}")

recommendations = get_recommendations(sample_product['name'])
if recommendations is not None:
    print("\nTop 5 Recommended Products:")
    display(recommendations[['name', 'main_category', 'sub_category', 'ratings', 'actual_price']])

# 12. Helper function to search for products
def search_products(keyword="", n=5):
    """
    Search for products by keyword in name or category

    Parameters:
    - keyword: Search term to look for in product names or categories
    - n: Maximum number of results to return
    """
    if not keyword:
        return df_sample[['name', 'main_category']].sample(min(n, len(df_sample)))

    keyword = keyword.lower()
    mask = (
        df_sample['name'].str.lower().str.contains(keyword, na=False) |
        df_sample['main_category'].str.lower().str.contains(keyword, na=False) |
        df_sample['sub_category'].str.lower().str.contains(keyword, na=False)
    )

    results = df_sample[mask].head(n)

    if len(results) == 0:
        print(f"No products found matching '{keyword}'. Here are some random products:")
        return df_sample[['name', 'main_category']].sample(min(5, len(df_sample)))

    return results[['name', 'main_category', 'sub_category', 'ratings', 'actual_price']]

# Example usage:
# search_products("phone")  # Search for products containing "phone"
# search_products("electronics")  # Search in categories
# search_products()  # Get random products
```
